**1.Draw a line in a diagram from position (1,3) to (2,10) then to (6,12) and finally to position (18,20).(mark each point with a beautiful color and set line color to red and line style dotted)**

```python
x=np.array([1,2,6,18])
y=np.array([3,10,12,20])
py.plot(x,y,marker='.',color='red',ms='15',mec= 'green',mfc='green',linestyle='dotted')
py.show()
```
