**1.Draw a line in a diagram from position (1,3) to (2,10) then to (6,12) and finally to position (18,20).(mark each point with a beautiful color and set line color to red and line style dotted)**

```python
x=np.array([1,2,6,18])
y=np.array([3,10,12,20])
py.plot(x,y,marker='.',color='red',ms='15',mec= 'green',mfc='green',linestyle='dotted')
py.show()
```

**7.create bar plot of scores by group and gender.use multiple X values on the same chart for men and women**
**sample data:means(men)=(22,30,35,26),means(women)=(25,32,30,35,29)**

```python
y1=[22,30,35,35,26]
y2=[25,32,30,35,29]
x=["g1","g2","g3","g4","g5"]
bw=0.2
xpos=np.arange(len(x))
py.bar(xpos-bw/2,y1,bw,label="men")
py.bar(xpos+bw/2,y2,bw,label="women")
py.xticks(xpos,x)
py.legend() 
py.xlabel("groups")
py.ylabel("scores")
py.show()
```

**10.write a python program to draw a scatter plot comparing two subject marks of mathematics and science. Use marks of 10 students.**
**test data: math_marks=[88,92,80,89,100,80,60,100,80,34], science_marks=[35,79,79,48,100,88,32,45,20,20],marks_range=[10,20,30,409,50,60,70,80,90,100]**
