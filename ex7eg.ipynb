```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
data={
    'sore_throat':[1,0,1,1,0,0,1,0,1,1],
    'fever':[1,0,0,0,1,0,0,1,1,1],
    'swollen_glans':[1,0,1,0,0,0,1,0,0,0],
    'congestion':[1,1,0,0,1,1,0,1,1,1],
    'headache':[1,1,0,0,0,0,0,1,1,1],
    'diagnosis':['Strep_Thorat','Allergy','Cold','Strep_Thorat','Cold','Allergy','Strep_Thorat','Allergy','Cold','Cold']
}
df=pd.DataFrame(data)
print(df)
```

```python
X=df[['sore_throat','fever','swollen_glans','congestion','headache']]
y=df['diagnosis']
clf=DecisionTreeClassifier(criterion="entropy")
clf.fit(X,y)
```

```python
plt.figure(figsize=(12,6))
plot_tree(clf,feature_names=X.columns,class_names=clf.classes_,filled=True,rounded=True,fontsize=10)
plt.show()
```

```python
newcase=[[0,0,1,0,0]]
prediction=clf.predict(newcase)
print("Prediction for (0,0,1,0,0):",prediction[0])
```
